```python
from imblearn.over_sampling import RandomOverSampler
ros = RandomOverSampler(random_state=0)
train_x, train_y = ros.fit_resample(np.array(train_data['sentence']).reshape(-1, 1), np.array(train_data['label']).reshape(-1, 1))
train = pd.DataFrame(list(zip([x[0] for x in train_x], train_y)), columns = ['sentence', 'label'])
```

```python
from sklearn import preprocessing
le = preprocessing.OneHotEncoder()
y_train= le.fit_transform(np.array(train['label']).reshape(-1, 1)).toarray()
y_test= le.fit_transform(np.array(test_data['label']).reshape(-1, 1)).toarray()
```

```python
from transformers import RobertaTokenizerFast
tokenizer = RobertaTokenizerFast.from_pretrained("roberta-base")
```

```python
def roberta_encode(data,maximum_length) :
  input_ids = []
  attention_masks = []
  

  for i in range(len(data.sentence)):
      encoded = tokenizer.encode_plus(
        
        data.sentence[i],
        add_special_tokens=True,
        max_length=maximum_length,
        pad_to_max_length=True,
        
        return_attention_mask=True,
        
      )
      
      input_ids.append(encoded['input_ids'])
      attention_masks.append(encoded['attention_mask'])
  return np.array(input_ids),np.array(attention_masks)
```

```python
max_len = max([len(x.split()) for x in train_data['sentence']])
train_input_ids,train_attention_masks = roberta_encode(train, max_len)
test_input_ids,test_attention_masks = roberta_encode(test_data, max_len)
```

```python
def create_model(bert_model, max_len, num_classes):
    input_ids = Input(shape=(max_len,), dtype='int32')
    attention_masks = Input(shape=(max_len,), dtype='int32')

    class BertLayer(tf.keras.layers.Layer):
        def __init__(self, bert_model, **kwargs):
            super(BertLayer, self).__init__(**kwargs)
            self.bert_model = bert_model

        def call(self, inputs):
            return self.bert_model(inputs)

    bert_output = BertLayer(bert_model)([input_ids, attention_masks])
    pooled_output = bert_output[1]

    output = Dense(num_classes, activation='softmax')(pooled_output)

    model = Model(inputs=[input_ids, attention_masks], outputs=output)
    model.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=1e-5), 
                  loss='categorical_crossentropy', 
                  metrics=['accuracy'])
    return model
```

```python
from transformers import TFRobertaModel
max_len = 128
num_classes = 2
bert_model = TFRobertaModel.from_pretrained("roberta-base")
```

```python
model = create_model(bert_model, max_len=max_len, num_classes=num_classes)
model.summary()
```

```python
history = model.fit([train_input_ids, train_attention_masks], y_train, epochs=10, batch_size=100)
```

```python
plt.plot(history.history['accuracy'])
plt.title('model accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')
plt.legend(['train'], loc='upper left')
plt.show()
```

```python
plt.plot(history.history['loss'])
plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['train'], loc='upper left')
plt.show()
```

```python
result = model.predict([test_input_ids,test_attention_masks])
y_pred = np.zeros_like(result)
y_pred[np.arange(len(result)), result.argmax(1)] = 1
```

```python
from sklearn.metrics import accuracy_score, f1_score,precision_score, recall_score
accuracy = accuracy_score(y_test, y_pred)
print('Accuracy ', accuracy)
f1 = f1_score(y_test, y_pred, average = 'macro')
print('F1 Score :', f1)
```

```python
from sklearn.metrics import accuracy_score, f1_score,precision_score, recall_score
for i in range(len(y_test[0])):
    accuracy = accuracy_score(y_test, y_pred)
    precision = precision_score(y_test[:, i], y_pred[:, i])
    recall = recall_score(y_test[:, i], y_pred[:, i])
    f1 = f1_score(y_test[:, i], y_pred[:, i])
    print(f"Label {i}: Accuracy={accuracy}, Precision={precision}, Recall={recall}, F1-score={f1}")
```
